```python
import yfinance as yf
```

```python
rel = yf.Ticker("RELIANCE.NS")
```

```python
rel = rel.history(period="max")
```

```python
 rel.plot.line(y="Close", use_index=True)
```

```python
if "Dividends" in rel:  
    del rel["Dividends"]  
```

```python
if "Stock Splits" in rel: 
    del rel["Stock Splits"]
```

```python
rel["Tomorrow"] = rel["Close"].shift(-1)
```

```python
rel
```

```python
rel["Target"] = (rel["Tomorrow"] > rel["Close"]).astype(int)
```

```python
rel
```

```python
rel = rel.loc["2010-01-01":].copy()
```

```python
rel
```

```python
from sklearn.ensemble import RandomForestClassifier
model= RandomForestClassifier(n_estimators=100, min_samples_split=100, random_state=1)

train = rel.iloc[:-100]
test = rel.iloc[-100:]

predictors = ["Close", "Open", "Volume", "High", "Low"] 
model.fit(train[predictors], train["Target"])  
```

```python
from sklearn.metrics import precision_score

preds = model.predict(test[predictors])
```

```python
import pandas as pd
preds = pd.Series(preds, index=test.index)
```

```python
precision_score(test["Target"], preds)
```

```python
combined = pd.concat([test["Target"], preds], axis=1)
```

```python
combined.plot()
```

```python
def predict(train, test, predictors, mdoel):
    model.fit(train[predictors], train["Target"])
    preds = model.predict(test[predictors])
    preds = pd.Series(preds, index= test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
def backtest(data, model, predictors, start=2500, step=250):  #250 days is one year for a stock so we are taking 10 yrs of data to backtest
    all_predictions = [] 

    for i in range(start, data.shape[0], step): 
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i+step)].copy() 
        predictions = predict(train, test, predictors, model)
        all_predictions.append(predictions)
    return pd.concat(all_predictions)
    
```

```python
predictions = backtest(rel, model, predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```

```python
predictions["Target"].value_counts() / predictions.shape[0]
```

```python
horizons = [2,5,60,250,1000]
new_predictors = [] 

for horizon in horizons: 
    rolling_averages = rel.rolling(horizon).mean() 

    ratio_column = f"Close_Ratio_{horizon}" 
    rel[ratio_column] = rel["Close"] / rolling_averages["Close"]

    trend_column = f"Trend_{horizon}" 
    rel[trend_column] = rel.shift(1).rolling(horizon).sum()["Target"] 

    new_predictors  += [ratio_column, trend_column] 
```

```python
rel
```

```python
model = RandomForestClassifier(n_estimators=200, min_samples_split=50, random_state=1) 
```

```python
def predict(train, test, predictors, mdoel):
    model.fit(train[predictors], train["Target"])
    preds = model.predict_proba(test[predictors])[:,1]
    preds[preds >= .6] = 1  
    preds[preds <  .6] = 0 
    preds = pd.Series(preds, index= test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
predictions = backtest(rel, model, new_predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```
